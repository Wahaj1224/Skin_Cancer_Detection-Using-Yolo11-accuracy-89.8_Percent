```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolo11n-cls.pt')  # load a pretrained model (recommended for training)


results = model.train(data='//content/drive/MyDrive/Downloads/archive', epochs=10, imgsz=224)


# Evaluate the model
results = model.val()

# Export the model to ONNX or other formats
success = model.export(format='onnx')
```
